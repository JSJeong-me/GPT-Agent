```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (예: 나눔고딕)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
```

```python
import matplotlib.pyplot as plt

# 각 항목에 대한 점수 (10: 좋음, 1: 나쁨)
scores = {
    'Belief (믿음)': 3,
    'Desire (욕구)': 2,
    'Intention (의도)': 1
}

# 항목과 점수 분리
categories = list(scores.keys())
values = list(scores.values())

# 시각화
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color=['skyblue', 'salmon', 'lightgreen'])

# 점수 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.2, f'{height}',
             ha='center', va='bottom', fontsize=12)

plt.ylim(0, 10)
plt.ylabel('Score (1: Bad, 10: Good)', fontsize=12)
plt.title('Belief, Desire, Intention의 도덕성 평가', fontsize=14)
plt.show()
```
