```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (예: 나눔고딕)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
```

```python
import matplotlib.pyplot as plt

# 주어진 내용에 대한 점수 (10: 좋음, 1: 나쁨)
# - Belief (신념): 자금 은닉 가능성에 대한 확신 → 2점
# - Desire (욕망): 범죄수익의 완전 은폐 및 위험 최소화 → 2점
# - Intention (의도): 가족 동원을 통한 체계적 실행 등 → 1점
scores = {
    "Belief (신념)": 2,
    "Desire (욕망)": 2,
    "Intention (의도)": 1
}

# 데이터 준비
categories = list(scores.keys())
values = list(scores.values())

# 시각화
plt.figure(figsize=(8,6))
bars = plt.bar(categories, values, color=['skyblue', 'lightgreen', 'salmon'])

# y축 범위 설정 (1점 ~ 10점)
plt.ylim(0, 11)

# 제목 및 축 레이블 추가
plt.title("범죄 자금 은닉 관련 요소의 도덕성 평가 (Good:10 - Bad:1)")
plt.ylabel("점수 (1: 나쁨, 10: 좋음)")

# 각 막대 위에 점수 값 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f'{yval}', ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# BDI 점수 정의 (0~10 척도)
belief_score = 8  # 신뢰 기반 설득
desire_score = 9  # 강한 금전적 이익 유도
intention_score = 8  # 행동 유도 강도

# 레이블과 점수 정의
labels = ["Belief (믿음)", "Desire (욕구)", "Intention (의도)"]
scores = [belief_score, desire_score, intention_score]

# 막대 그래프 생성
plt.figure(figsize=(8, 5))
plt.bar(labels, scores, color=['blue', 'orange', 'red'])

# 점수 표시
for i, score in enumerate(scores):
    plt.text(i, score + 0.3, str(score), ha='center', fontsize=12, fontweight='bold')

# 그래프 제목 및 축 설정
plt.ylim(0, 10)
plt.title("BDI 분석을 통한 보이스 피싱 가능성 평가", fontsize=14, fontweight='bold')
plt.ylabel("점수 (0~10)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프 출력
plt.show()
```
