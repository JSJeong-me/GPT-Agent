```python
import json

def read_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    return data

def analyze_json(data, indent=0):
    spacing = ' ' * indent
    if isinstance(data, dict):
        for key, value in data.items():
            print(f"{spacing}Key: {key}, Type: {type(value).__name__}")
            analyze_json(value, indent + 4)
    elif isinstance(data, list):
        print(f"{spacing}List of {len(data)} items, Type: list")
        if data:
            analyze_json(data[0], indent + 4)
    else:
        print(f"{spacing}Value: {data}, Type: {type(data).__name__}")
```

```python
if __name__ == '__main__':
    file_path = 'voice_phishing_scam.json'  # JSON 파일 경로 입력
    json_data = read_json(file_path)
    analyze_json(json_data)
```

```python
from openai import OpenAI
import numpy as np

client = OpenAI()

def normalize_l2(x):
    x = np.array(x)
    if x.ndim == 1:
        norm = np.linalg.norm(x)
        if norm == 0:
            return x
        return x / norm
    else:
        norm = np.linalg.norm(x, 2, axis=1, keepdims=True)
        return np.where(norm == 0, x, x / norm)


response = client.embeddings.create(
    model="text-embedding-3-small", input="Testing 123", encoding_format="float"
)

cut_dim = response.data[0].embedding[:256]
norm_dim = normalize_l2(cut_dim)

print(norm_dim)
```
