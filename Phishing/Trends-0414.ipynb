```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 정의: 각 범주는 "전체" 또는 "1월~7월" 기간 구분 추가
categories = ["2019 전체", "2023 전체", "2023 (1월~7월)", "2024 (1월~7월)"]

# 피해금액 (단위: 억 원)
damage_amount = [6398, 4472, 2427, 3909]

# 피해건수 (건수 수치: 천 단위 구분 없이 정수)
damage_cases = [37667, 18902, 1741, 11734]

# x축 위치 설정
x = np.arange(len(categories))

# 그래프 그리기: 2행 1열 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# 피해금액 그래프
ax1.bar(x, damage_amount, color='cornflowerblue')
ax1.set_title("연도별 피해금액 (단위: 억 원)")
ax1.set_xticks(x)
ax1.set_xticklabels(categories, rotation=45)
ax1.set_ylabel("피해금액 (억 원)")
# 막대 위에 값 표시
for idx, value in enumerate(damage_amount):
    ax1.text(x[idx], value + 50, str(value), ha='center', va='bottom', fontsize=9)

# 피해건수 그래프
ax2.bar(x, damage_cases, color='seagreen')
ax2.set_title("연도별 피해건수")
ax2.set_xticks(x)
ax2.set_xticklabels(categories, rotation=45)
ax2.set_ylabel("피해건수")
# 막대 위에 값 표시
for idx, value in enumerate(damage_cases):
    ax2.text(x[idx], value + 200, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 정의: 각 범주는 "전체" 또는 "1월~7월" 기간 구분 추가
categories = ["2019 전체", "2023 전체", "2023 (1월~7월)", "2024 (1월~7월)"]

# 피해금액 (단위: 억 원)
damage_amount = [6398, 4472, 2427, 3909]

# 피해건수
damage_cases = [37667, 18902, 1741, 11734]

# x축 위치 설정
x = np.arange(len(categories))
width = 0.5  # 막대 폭을 줄였습니다.

# 그래프 그리기: 2행 1열 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# 피해금액 그래프
ax1.bar(x, damage_amount, width=width, color='cornflowerblue')
ax1.set_title("연도별 피해금액 (단위: 억 원)")
ax1.set_xticks(x)
ax1.set_xticklabels(categories, rotation=45)
ax1.set_ylabel("피해금액 (억 원)")
# 막대 위에 값 표시
for idx, value in enumerate(damage_amount):
    ax1.text(x[idx], value + 50, str(value), ha='center', va='bottom', fontsize=9)

# 피해건수 그래프
ax2.bar(x, damage_cases, width=width, color='seagreen')
ax2.set_title("연도별 피해건수")
ax2.set_xticks(x)
ax2.set_xticklabels(categories, rotation=45)
ax2.set_ylabel("피해건수")
# 막대 위에 값 표시
for idx, value in enumerate(damage_cases):
    ax2.text(x[idx], value + 200, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
